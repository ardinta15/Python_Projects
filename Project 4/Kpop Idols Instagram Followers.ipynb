# Data Analysis

#### Gender Distribution

```python
Gender_counts = idol['Gender'].value_counts()
Gender_counts.plot(kind='pie', autopct='%6.2f%%', colors=['green', 'magenta'], labels=['Male', 'Female'])
plt.title('Gender Distribution')
plt.show()

male_idol = idol[idol.Gender == 'M']
female_idol = idol[idol.Gender == 'F']

males, attributes = male_idol.shape
print(f'There are {males} male idols in the dataset.')
females, attributes = female_idol.shape
print(f'There are {females} female idols in the dataset.')
```

#### Top 5 Idols Most Followed

```python
top_5_followed = idol.nlargest(5, 'Followers')
top_5_followed.plot(x = 'Stage Name', y = 'Followers', kind = 'bar', legend = False)


plt.xlabel('Stage Name')
plt.ylabel('Followers')
plt.title('Top 5 Idols Most Followed')


plt.show()
```

```python
top_5 = top_5_followed['Stage Name'].tolist()
print(f"The top 5 followed performers are: {', '.join(top_5)}.")
```

#### Top 5 Groups Most Followed

```python
top_5_followed = idol.groupby('Group')['Followers'].sum().nlargest(5)
top_5_followed.plot(x = 'Group', y = 'Followers', kind = 'bar', legend = False)


plt.xlabel('Group')
plt.ylabel('Followers')
plt.title('Top 5 Groups Most Followed')


plt.show()
```
